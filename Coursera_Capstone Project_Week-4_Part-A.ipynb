## **Title: Restaurant Recommender System for Chennai**

### **Problem background:**
**Chennai** (previously **Madras**) is the capital of the Indian state of Tamil Nadu. It is one of the largest cultural, economic and educational centre of south India. It is the fourth-most populous urban agglomeration in India. Chennai is nicknamed "The Detroit of India", with more than one-third of India's automobile industry being based in the city. 


Chennai is among the most-visited Indian cities by foreign tourists. Chennai attracts 45 percent of health tourists visiting India, and 30 to 40 percent of domestic health tourists. As such, it is termed "India's health capital".


Chennai had the third-largest expatriate population in India, estimated at over 100,000 by 2016. National Geographic mentioned Chennai as the only South Asian city to feature in its 2015 "Top 10 food cities" list.

### **Problem description:**
Chennai is the home for large number of non-local IT employees and employees working in automobile sectors and tourist and health visitors. Since good food is an important factor for everyone. Speciality food also attract people around the world to try it out. In such scenarios, we need to find the right place, based on our budget. 


There are few questions that must be addressed,  
1.	Variety of foods which are available in the restaurant?
2.	Cost for a couple in this restaurant? 
3.	Nearby Restaurant with good rating?
4.	Similar restaurants nearby to me?
5.	Cost for a couple in similar restaurants? 


To address these questions, we need to good Restaurant recommender system which uses FourSquare location mapping the key localities in Chennai.


### **Target audience:**
Target audiences who uses this Restaurant Recommender app belong to both local and non-local people staying in Chennai. Also tourists and Visitors who stay in Chennai for few days.


People can select restaurants nearby and can look for a similar restaurant which caters specific cuisine of food. 


People who rarely use restaurants would prefer the most rated restaurants. 
People who regularly have food in restaurants would prefer search for new restaurants in the city and try new varieties of food. 


Hence target audience is everyone who is exploring different places or similar cuisines.
